```
from PIL import Image
import requests
from io import BytesIO
from transformers import pipeline

# Carregar modelo pré-treinado de classificação de imagem
classificador = pipeline("image-classification", model="microsoft/resnet-50")

# URL da imagem de exemplo (caso a anterior tenha problemas)
url_imagem = "https://upload.wikimedia.org/wikipedia/commons/9/9a/Pug_600.jpg"

# Cabeçalhos para simular uma requisição de navegador
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Fazer requisição com cabeçalhos
response = requests.get(url_imagem, headers=headers)

# Verificar o status da resposta e o tipo de conteúdo
if response.status_code == 200:
    # Verificar se o conteúdo retornado é uma imagem
    if 'image' in response.headers['Content-Type']:
        try:
            # Abrir a imagem com Pillow
            imagem = Image.open(BytesIO(response.content))
            imagem.show()

            # Classificar a imagem usando o modelo pré-treinado
            resultados = classificador(imagem)

            # Mostrar os resultados da classificação
            print("Resultados da classificação:\n")
            for r in resultados:
                print(f"{r['label']} - Confiança: {r['score']:.2%}")
        except Exception as e:
            print(f"Erro ao abrir a imagem: {e}")
    else:
        print("O conteúdo retornado não é uma imagem.")
else:
    print(f"Erro ao carregar a imagem. Status Code: {response.status_code}")
```
